```python
df = (
    df.drop(columns=['updated_at', 'moderated_at', 'send_at'])
    .rename(columns={"id": "cash_request_id"})
)
df
```

```python
df_combined = pd.merge(df, df_fees, on='cash_request_id', how='outer')
df_combined
```

```python
df_combined.info()
```

```python
df_combined = (df_combined
    .drop(columns=['created_at_y', 'status_y', "reason", "updated_at", "from_date", "to_date", "recovery_status", "reco_last_update", "reco_creation"])
    .rename(columns={"status_x": "status", "created_at_x": "created_at", "id": "fees_id", "reimbursement_date": "expected_date", "type": "fee_type", "category": "incident_category", "total_amount": "fee_amount"}))
df_combined
```

```python
df_combined.info()
```

```python
df_combined = df_combined.dropna(subset=['cash_request_id'])
df_combined
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_combined["fee_type"] = df_combined["fee_type"].fillna("cash_request")
df_combined
```

```python
nan_users = df_combined[df_combined['user_id'].isna()]
nan_users
```

```python
# Rellenar los valores NaN de 'user_id' con los valores de 'deleted_account_id'
df_combined['user_id'] = df_combined['user_id'].fillna(df_combined['deleted_account_id'])

# Verificar si los valores vacos en 'user_id' han sido rellenados
print(df_combined[['user_id', 'deleted_account_id']])
```

```python
cash_requests = df_combined[df_combined['fee_type'] == 'cash_request']

# Filtrar los usuarios que tienen ms de un 'cash_request'
repeated_users = cash_requests.groupby('user_id').filter(lambda x: len(x) > 1)

# Mostrar el DataFrame filtrado
repeated_users
```

```python
print(repeated_users['user_id'].nunique())
```

```python
repeated_users['created_at'] = pd.to_datetime(repeated_users['created_at'])

# Ordenar por 'user_id' y 'created_date' para calcular las diferencias
repeated_users = repeated_users.sort_values(by=['user_id', 'created_at'])

# Calcular la diferencia de tiempo entre cada 'cash_request' por usuario
repeated_users['time_diff'] = repeated_users.groupby('user_id')['created_at'].diff()

# Mostrar el DataFrame con la nueva columna 'time_diff' que contiene la diferencia de tiempo
repeated_users[['user_id', 'created_at', 'time_diff']]
```

```python
repeated_users = repeated_users['time_diff'].fillna(0, inplace=True)
```

```python
repeated_users
```

```python
# Asegrate de que 'time_diff' est en formato timedelta
# Convertir la columna 'time_diff' a das
repeated_users['time_diff_days'] = repeated_users['time_diff'].dt.days

# Mostrar el DataFrame con la nueva columna
print(repeated_users[['user_id', 'created_at', 'time_diff_days']])
```
